**PROJETINHO DE ESTATÍSTICA**


$$ x \sim N(\mu, \sigma^2) -> P(|x - \mu| \leq k\sigma)) $$

```python
p1 = ss.norm.cdf(1) - ss.norm.cdf(-1) # Probabilidade pra K=1...
p2 = ss.norm.cdf(2) - ss.norm.cdf(-2)
p3 = ss.norm.cdf(3) - ss.norm.cdf(-3)
print('P(|X| < 1) = {}'.format(round(p1, 4)))
print('P(|X| < 2) = {}'.format(round(p2, 4)))
print('P(|X| < 3) = {}'.format(round(p3, 4)))
#Resultado da probabilidade da normal 1 padrão e do desvio padrão... e normal 2...
```
